```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def function(x):
    return (x + 3) ** 2

def derivative(x):
    return 2 * (x + 3)

# Gradient Descent parameters
learning_rate = 0.1  # Step size
initial_x = 2        # Starting point
precision = 0.0001   # Stop when the gradient is smaller than this
max_iterations = 100 # Maximum number of iterations

# Lists to store the x and y values for visualization
x_values = [initial_x]
y_values = [function(initial_x)]

# Gradient Descent algorithm
x = initial_x
for i in range(max_iterations):
    gradient = derivative(x)
    new_x = x - learning_rate * gradient
    x_values.append(new_x)
    y_values.append(function(new_x))
    
    # Stop if the change in x is less than the precision
    if abs(new_x - x) < precision:
        break
    
    # Update x to the new x value
    x = new_x

# Display the result
print(f"Local minimum occurs at x = {x:.4f}")
print(f"Minimum value of function is y = {function(x):.4f}")

# Plotting the function and the gradient descent path
x_range = np.linspace(-6, 4, 100)
y_range = function(x_range)

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label="y = (x + 3)^2", color="blue")
plt.scatter(x_values, y_values, color="red", label="Gradient Descent Path")
plt.plot(x_values, y_values, linestyle="--", color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent to find local minima of y = (x + 3)^2")
plt.legend()
plt.grid()
plt.show()
```
